```python
import pandas as pd
import numpy as np
df =pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',
                   header=0,skiprows=0)
df=df[0]
df=df[df['Borough']!='Not assigned']
df2=pd.read_csv('http://cocl.us/Geospatial_data')
df=df.sort_values('Postcode')
df=pd.DataFrame.reset_index(df,level=None)
del df['index']
Pcode=df['Postcode']
post=df2['Postal Code']
Lat=df2['Latitude']
Lon=df2['Longitude']
p=[]
q=[]
j=0
for i in range(0,212):
    while(j<105):
        if(Pcode[i]==post[j]):
            p.append(Lat[j])
            q.append(Lon[j])
            break
            j=j-1
        j=j+1
df['Latitude']=p
df['Longitude']=q
df=df.sort_values('Neighbourhood')
df=pd.DataFrame.reset_index(df,level=None)
del df['index']
df
```
